```python
from pyspark.sql import SparkSession

# Initiate SparkSession
spark = SparkSession.builder \
            .appName('CSV Handling in Pyspark') \
            .getOrCreate()

print("Spark is ready")
```

```python
# Basic CSV Read
df = spark.read.csv("./dataset/employee_file.csv")
df.show(5)
```

```python
# Reading CSV with options

# Header: if your CSV file has a header row, set header=True to use it as column names.
# InferSchema: To automatically determine the data types of each column, set inferSchema=True
# Delimiter: To use custom delimiter such as `,`, `;` `|` set the sep option.


df = spark.read.option("header", True) \
                .option("inferSchema", True) \
                .option("sep", ",") \
                .csv("./dataset/employee_file.csv")

df.show(5)
```

```python
# Additional Reading Options

# Null Values: Use nullValue to specify a string to interpret as null, such as "N/A" or "NULL".
# Date Format: Set dateFormat to specify the date format for date fields
# Mode : Determine how Spark handles corrupt records.
    # PERMISSIVE (default): Puts corrupt records in a separate column.
    # DROPMALFORMAT: Drops rows with bad records
    # FAILFAST: Throws an error for corrupt records
```

```python
df = spark.read.option("header", True) \
                .option("inferSchema", True) \
                .option("nullValue", "N/A") \
                .option("dateFormat", "YYYY-mm-dd") \
                .option("mode", "DROPMALFORMED") \
                .csv("./dataset/employee_file.csv")
df.show(5)
```

```python
# Basic CSV write
# df.write.csv("/path/to/output_data")
```

```python
# PySpark gives several options for customizing how the output is saved.

# Header: Set header=True to include the column names in the first row
# Mode: Controls the bahavior if the output file or directory already exists.
    # overwrite: Replace existing date.
    # append: Adds data to existing file
    # ignore: Skips writing if the file exists
    # error or errorifexists: Throws an error if the file exists.
```

```python
df.write.option("header", True) \
        .mode("overwrite") \
        .csv("./dataset/output_data/employee_data")
```
